# **1. What is the difference between a function and a method in Python?**
# **answer**
Function: A function is a block of reusable code that performs a specific task. It can be defined outside of a class and is called using its name.

Example:



def my_function():

    return "Hello"
Method: A method is a function that is associated with an object and is defined inside a class. It is called on an instance (object) of the class and typically operates on that object.

Example:


class MyClass:

    def my_method(self):

        return "Hello"

obj = MyClass()

obj.my_method()  # Calling method on object

#**5. What are iterators in Python and how do they differ from iterables?**

#**answer**
Iterable: An iterable is an object that can be looped over (iterated) using a loop like for. It implements the __iter__() method and can return an iterator.

Examples of iterables: lists, tuples, dictionaries, strings.
Iterator: An iterator is an object that represents a stream of data and keeps track of the current state. It implements the __next__() method and can be used to iterate over the elements one by one.

Example:


my_list = [1, 2, 3]

iterator = iter(my_list)

print(next(iterator))  # Output: 1


#**6. Explain the concept of generators in Python and how they are defined.**
#**answer**
A generator is a special type of iterator that generates values lazily, meaning it produces values on the fly when requested instead of storing them in memory. Generators are defined using functions with yield statements.

Example:


def my_generator():
    yield 1
    yield 2
    yield 3


gen = my_generator()

print(next(gen))  # Output: 1

print(next(gen))  # Output: 2


#**7. What are the advantages of using generators over regular functions?**
#**answer**
Memory Efficiency: Generators produce values one at a time, so they don’t require memory to store the entire output at once, which is especially useful for large datasets.

Lazy Evaluation: Values are computed only when needed, improving performance when iterating through large sequences.

Cleaner Code: Generators provide a more compact way to implement iterators compared to writing explicit classes with __iter__() and __next__() methods.

#**8. What is a lambda function in Python and when is it typically used?**
#**answer**
A lambda function is an anonymous (unnamed) function defined with the lambda keyword. It can have any number of arguments but only one expression. It's often used for short-term operations where defining a full function is not necessary.

Example:


add = lambda a, b: a + b

result = add(5, 3)  # Output: 8


Typical use cases:

When a simple function is required for a short period, e.g., passing it as an argument to map(), filter(), or sorted() functions.


#**9. Explain the purpose and usage of the map() function in Python.**
#**answer**

The map() function applies a given function to all items in an iterable (like a list or tuple) and returns an iterator that yields the results. It's useful for transforming data.



Example:


def square(x):

    return x * x

numbers = [1, 2, 3, 4]

result = map(square, numbers)

print(list(result))  # Output: [1, 4, 9, 16]
#**10. What is the difference between map(), reduce(), and filter() functions in Python?**
#**answer**

map(): Applies a function to each item of an iterable and returns an iterator with the results.

Example: map(function, iterable)
reduce() (from functools module): Applies a function cumulatively to the items of an iterable, reducing it to a single value (e.g., summing all elements).

Example:


from functools import reduce

result = reduce(lambda x, y: x + y, [1, 2, 3, 4])

print(result)  # Output: 10


filter(): Applies a function that returns a boolean value to each item of an iterable and filters out elements for which the function returns False.

Example: filter(function, iterable)


Summary:

map(): Transforms each element.

reduce(): Reduces all elements to a single value.

filter(): Filters elements based on a condition.
